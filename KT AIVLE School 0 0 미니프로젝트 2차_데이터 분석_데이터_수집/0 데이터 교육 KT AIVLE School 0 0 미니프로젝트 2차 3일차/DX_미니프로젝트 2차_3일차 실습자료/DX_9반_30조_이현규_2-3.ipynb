## 3. 이변량 분석

- 수치형 feature --> 수치형 target
    - 전체 변수들 간의 상관관계 구하기
        * 범주형 변수를 제외한 데이터셋을 이용하여
        * .corr() + sns.heatmap() 으로 전체 상관계수를 시각화
    - 상관계수 상위 몇개에 대해서 feature와 target에 대해 
        * 상관분석을 통해 상관계수가 유의미함을 분석
        * 산점도를 통해 상관관계를 시각화하여 분석
- 범주형 feature --> 수치형 target
    * 범주간에 target의 평균의 차이가 있는지 분석
    * 범주별 barplot으로 평균의 차이를 시각화하여 분석

## [정리] 새로운 Target에 대한 데이터 분석을 통해 파악된 비즈니스 인사이트는 무엇인가요?

**공용주차장 건립을 위한 분석**

상가와 주택이 증가하고 세대당 등록차량비율이 증가함에 따라 주차장이 부족한 문제가 생겼고 공공기관에서 주차장 건립을 위한 분석을 의뢰했다.

일단 이전 데이터 분석에서는 등록차량수에 대한 상관계수 분석을 했을 때에는 '총세대수','전용면적_40_50','전용면적_50_100','임대료','임대보증금'
부분에서 등록차량수와 가장 상관관계가 있는 것으로 분석했다.
하지만 등록차량수만을 놓고 주차장 건립에 대해서 분석하기에는 총세대수가 많아 차량의 대수가 많다고 주차장을 우선적으로 건립할 수도 없다.
그래서 우리는 등록차량수비율을 가지고 분석을 했다. 등록차량수비율을 가지고 분석을 하니 등록차량수로만 비교를 했을 때와 상관관계적을 다른 분석내용이 나왔다.


주차장의 건립에 있어서 과연 더 중요한 것이 등록차량수인지 등록차량수 비율인지 알아봐야 될 것 같다. 일단 상가 지역보다 아파트 지역에 사람들의 인구 밀도가 더 몰려있다.

```python
# 2일차 데이터에서 1대당 주차면수 = 주차면수 / 주차차량수 해서 새로운 칼럼을 추가
# data['1대당주차면수'] = data['단지내주차면수'] / data['등록차량수']
# data.head()


target = '등록차량수_비율'
feature = '1대당주차면수'
print(f"[{feature}] 통계 분석 및 그래프 분석")

# 통계 분석 : 통계 분석

print("   ***** 통계 분석 *****")

## 코드 입력
result = spst.pearsonr(re_data[feature],re_data[target])
    
print(feature, " vs ", target, " 상관 분석: ", result)
    
if result[1] > 0.05:
        print(f"통계분석 결과 : {feature}는 등록차량수 비율에 영향을 주지 않는다")
else:
        print(f"통계분석 결과 : {feature}는 등록차량수 비율에 영향을 준다")

# 그래프 분석 : regplot

# plt.figure(figsize = (12,8))
print("   ***** 그래프 분석 *****")
    
## 코드 입력
sns.regplot(x = feature , y = target , data = re_data)
plt.grid()
plt.show()
    
print("")
print("-"*50)
```
